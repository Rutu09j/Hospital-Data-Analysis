# 5. Daily Average Medical Expenses 
#Calculate the average medical expenses per day for each hospital.

```python
df.columns
```

```python
df['number_of_days'] = (df['Discharge_Date'] - df['Admission_Date']).dt.days+1
```

```python
df['expense_per_day'] =df['Medical_Expenses']/df['number_of_days']
```

```python
result = df.groupby('Hospital_Name')['expense_per_day'].mean().reset_index()
```

```python
result.rename(columns={'expense_per_day': 'average_expense_per_day'} , inplace =True)
```

```python
print(result)
```

```python
plt.figure(figsize=(16,10))
ax =sns.barplot(x ="Hospital_Name" , y= "average_expense_per_day" , data=result, color="Green")
plt.title("Daily Average Expense per Hospital")
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel("Hospital_Name")
plt.ylabel("average_expense_per_day ")
plt.show()
```

#  6) Longest Hospital Stay 
#Find the patient with the longest stay by calculating the difference between 
 #Discharge Date and Admission Date.

```python
df['stay_duration'] = (df['Discharge_Date'] - df['Admission_Date']).dt.days+1
longest_stay_patient = df.loc[df['stay_duration'].idxmax()]
print("Patient with the longest stay:")
print(longest_stay_patient)
```

```python
plt.bar('longest_stay_patient' ,longest_stay_patient['stay_duration'],color = 'skyblue')
plt.title("Patients with longest stay")
plt.ylabel("Number of Days")
plt.show()
```

 #  8. Average Length of Stay Per Department 
 #Calculate the average number of days patients spend in each department.

```python
df['stay_duration'] = (df['Discharge_Date'] - df['Admission_Date']).dt.days + 1
avg_stay = df.groupby('Department')['stay_duration'].mean().reset_index()
avg_stay_columns = ['Department' , 'average_stay_days']
print(avg_stay)
```

#  10. Monthly Medical Expenses Report 
 #Group the data by month and calculate the total medical expenses for each month.

```python
df['Month_Year'] = df['Admission_Date'].dt.to_period("M")
monthly_expenses = df.groupby('Month_Year')['Medical_Expenses'].sum().reset_index()
print(monthly_expenses)
```

```python
plt.figure(figsize=(16,10))
plt.plot(monthly_expenses['Month_Year'].astype(str) , monthly_expenses['Medical_Expenses'], marker= 'o', color= 'brown')
plt.title("Monthly Medical Expenses Report (based on Admission date)")
plt.xlabel("Month")
plt.ylabel("Total Medical Expenses")
plt.show()
          
```
